$$

$$
U(\mu,\sigma^2) = a \mu - \frac{b}{2} \sigma^2
$$

where $a$ and $b$ are two positive scalars that describe the person's attitude toward bearing risk associated with a random variable with mean $\mu$ and variance $\sigma^2$.
